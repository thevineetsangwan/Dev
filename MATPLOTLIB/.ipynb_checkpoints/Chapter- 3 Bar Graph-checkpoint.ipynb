# Bar Plot

```python
import matplotlib.pyplot as plt

x = ['Python','Java','C++','C']
y = [85,70,60,45]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)
plt.title("Popularity",fontsize=20)

c = ['black','grey','blue','y']

plt.bar(x,y,width=0.3,color=c,align='edge',edgecolor='green',linewidth=2,linestyle=":",alpha=0.4,label='%')
plt.legend()

plt.show()
```

```python
z = [70,60,50,30]

# plt.zlabel("In Use",fontsize=15) -- it throws error that zlabel is not part of matlpotlib
plt.title("Bar Graph",fontsize=20)


plt.xlabel("Language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)

plt.bar(x,y,width=0.4,color='black',align='center',label='Popularity')
plt.bar(x,z,width=0.4,color='y',align='center',label='In Use')
plt.legend()

plt.show()
```

```python
import numpy as np

z = [70,60,50,30]

# plt.zlabel("In Use",fontsize=15) -- it throws error that zlabel is not part of matlpotlib
plt.title("Bar Graph",fontsize=20)


plt.xlabel("Language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)

plt.bar(x,y,width=0.4,color='black',align='center',label='Popularity')
plt.bar(x,z,width=0.4,color='y',align='center',label='In Use')
plt.legend()

plt.show()
```

```python
import numpy as np

z = [70,60,50,30]
p = np.arange(len(x)) # will create an array Like: [0,1,2,3]
print(p)
     
width = 0.2

p1 = [j+width for j in p] # adding 0.2 in each of the element of p (new to me)
print(p1)
     
# plt.zlabel("In Use",fontsize:15) -- it throws error that zlabel is not part of matlpotlib
plt.title("Bar Graph",fontsize=20)
plt.xlabel("Language" ,fontsize=15)
plt.ylabel("Percentage" ,fontsize=15)
     
plt.bar(p,y,width, color= 'black' ,align= 'center' , label= 'Popularity') #replace to p in place of x
plt.bar(p1,z,width, color= 'y' ,align= 'center' , label= 'In Use') #replace to p1 in place of x
        
plt.xticks(p+width/2,x,rotation=30) # 1/2 is to place it in between,show name under language, and rotation to rotate name to a degree 
plt.legend()
plt.show()
```

<h1> Horizontal Bar Graph </h1>

```python
import numpy as np

z = [70,60,50,30]
p = np.arange(len(x)) # will create an array Like: [0,1,2,3]
print(p)
     
width = 0.2

p1 = [j+width for j in p] # adding 0.2 in each of the element of p (new to me)
print(p1)
     
# plt.zlabel("In Use",fontsize:15) -- it throws error that zlabel is not part of matlpotlib
plt.title("Bar Graph",fontsize=20)
plt.xlabel("Language" ,fontsize=15)
plt.ylabel("Percentage" ,fontsize=15)
     
plt.barh(p,y,width, color= 'black' ,align= 'center' , label= 'Popularity') #replace to p in place of x
plt.barh(p1,z,width, color= 'y' ,align= 'center' , label= 'In Use') #replace to p1 in place of x
        
plt.xticks(p+width/2,x,rotation=30) # 1/2 is to place it in between,show name under language, and rotation to rotate name to a degree 
plt.legend()
plt.show()
```
